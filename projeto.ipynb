Conjunto de dados sobre qualidade de vinhos da UCI Machine Learning Lab (https://archive.ics.uci.edu/ml/datasets/Wine+Quality)

- Propriedades físico-químicas

    - 4	Açúcar residual
    - 5	Cloretos
    - 6	Dióxido de enxofre livre
    - 7	Dióxido de enxofre total
    - 8	Densidade
    - 9	pH
    - 10 Sulfatos
    - 11 Álcool
- Avaliação de qualidade
    - 12	Qualidade - Pontuação entre 0 e 10 (mediana de pelo menos 3 avaliações feitas por especialistas em vinho)

### Groupby

É possível estratir estatísticas com funções como describe ou mean, mas é genérico(ex: média do ph - aplicamos wine_df.mean() e conferimos a média dele)
Para saber grupos específicos (ex: média do ph por qualidade) podemos usar groupby

```python
wine_df.groupby('quality').mean()
```

```python
#ou ainda mais específico, dividindo por qualidade e cor
wine_df.groupby(['quality', 'color']).mean()
```

```python
#só uma coluna, sem cor e qualidade no índice
wine_df.groupby(['quality', 'color'], as_index=False)['pH'].mean()
```

- P2: Qual nível de acidez (valor de pH) recebe a classificação média mais alta?

Essa pergunta é mais complicada porque, ao contrário da cor, que possui categorias claras pelas quais você pode agrupar (tinto ou branco), pH é uma variável quantitativa, sem categorias claras. No entanto, você pode criar uma variável categórica de uma variável quantitativa criando suas próprias categorias. 

Será necessário criar categorias com níveis de acidez:
Alto: Abaixo de 25% dos valores de pH
Moderadamente alto: 25% a 50% dos valores de pH
Médio: 50% a 75% dos valores de pH
Baixo: 75% ou mais dos valores de pH

```python
#observe as estatíticas descritivas do pH - coloque em uma variável
ph_desc = wine_df['pH'].describe()
ph_desc
```

```python
#montar uma lista com os valores encontrados
ph_borda = [ph_desc['min'], ph_desc['25%'], ph_desc['50%'], ph_desc['75%'], ph_desc['max']]

#dar rótulos aos níveis numa lista
ph_names = ["Alto" ,"Moderadamente Alto" ,"Médio" ,"Baixo"]
```

```python
#criar nova coluna com os níveis de acidez (acidity levels)
#pd.cut
wine_df['acidity levels'] = pd.cut(wine_df['pH'], ph_borda, labels=ph_names)
wine_df.head()
```

```python
# Encontre a qualidade média de cada nível de acidez com groupby
wine_df.groupby(["acidity levels"]).mean()["quality"]
```

### Query

- Vinhos com maior teor alcoólico recebem classificações maiores?

```python
#valor mediano de teor alcoolico
alcool_median = wine_df['alcohol'].median()
alcool_median
```

```python
#pra usar variável em uma query usa-se @

#vamos criar amostras de baixo e alto teor alcoolico
baixo_alcool = wine_df.query('alcohol < @alcool_median')
alto_alcool = wine_df.query('alcohol >= @alcool_median')
```

```python
# certifique-se que estas consultas incluíram cada amostra uma única vez
num_amostras = wine_df.shape[0]
num_amostras == baixo_alcool['quality'].count() + alto_alcool['quality'].count() # resultado deve ser True
```

```python
baixo_alcool['quality'].mean()
```

```python
alto_alcool['quality'].mean()
```

- Vinhos mais doces (mais açúcar residual) recebem classificações maiores?

```python
##valor mediano de açucar
acucar_median = wine_df['residual sugar'].median()
acucar_median
```

```python
#vamos criar amostras de baixo e alto teor de açúcar
    #não vou usar query pq o nome tem espaçamento
baixo_acucar = wine_df[wine_df['residual sugar'] < acucar_median]
alto_acucar = wine_df[wine_df['residual sugar'] >= acucar_median]
```

```python
baixo_acucar['quality'].mean()
```

```python
alto_acucar['quality'].mean()
```
